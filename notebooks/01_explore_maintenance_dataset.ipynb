```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("‚úÖ Libraries imported!")
```

```python
# Load dataset
df = pd.read_csv('../data/raw/maintenance/ai4i2020.csv')

print(f"‚úÖ Dataset loaded!")
print(f"Shape: {df.shape[0]} rows √ó {df.shape[1]} columns")
```

```python
# Display first rows
df.head()
```

```python
# Dataset info
print("=== DATASET INFO ===")
df.info()
```

```python
# List all columns
print("=== COLUMNS ===\n")
for i, col in enumerate(df.columns, 1):
    print(f"{i:2d}. {col}")
```

```python
# Descriptive statistics
df.describe()
```

```python
# Check for missing values
print("=== MISSING VALUES ===\n")
nulls = df.isnull().sum()
print(nulls)
print(f"\n‚úÖ Total missing: {nulls.sum()}")
```

```python
# Analyze target variable
print("=== TARGET VARIABLE: Machine failure ===\n")
print(df['Machine failure'].value_counts())
print(f"\nüìä Failure rate: {df['Machine failure'].mean()*100:.2f}%")

# Visualize
df['Machine failure'].value_counts().plot(kind='bar')
plt.title('Machine Failure Distribution')
plt.xlabel('Failure (0=No, 1=Yes)')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# Analyze failure types
failure_cols = ['TWF', 'HDF', 'PWF', 'OSF', 'RNF']

print("=== FAILURE TYPES ===\n")
for col in failure_cols:
    count = df[col].sum()
    pct = count / len(df) * 100
    print(f"{col}: {count:3d} ({pct:.2f}%)")

# Visualize
failure_counts = [df[col].sum() for col in failure_cols]
plt.bar(failure_cols, failure_counts)
plt.title('Failure Types Distribution')
plt.xlabel('Failure Type')
plt.ylabel('Count')
plt.show()
```

```python
# Temperature distributions
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Air temperature
axes[0].hist(df['Air temperature [K]'], bins=50, edgecolor='black')
axes[0].set_title('Air Temperature')
axes[0].set_xlabel('Temperature (K)')

# Process temperature
axes[1].hist(df['Process temperature [K]'], bins=50, edgecolor='black')
axes[1].set_title('Process Temperature')
axes[1].set_xlabel('Temperature (K)')

# Scatter
axes[2].scatter(df['Air temperature [K]'], 
                df['Process temperature [K]'], 
                alpha=0.3)
axes[2].set_title('Air vs Process Temp')
axes[2].set_xlabel('Air (K)')
axes[2].set_ylabel('Process (K)')

plt.tight_layout()
plt.show()
```

```python
# RPM, Torque, Tool Wear
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

axes[0].hist(df['Rotational speed [rpm]'], bins=50, edgecolor='black')
axes[0].set_title('Rotational Speed')
axes[0].set_xlabel('RPM')

axes[1].hist(df['Torque [Nm]'], bins=50, edgecolor='black')
axes[1].set_title('Torque')
axes[1].set_xlabel('Nm')

axes[2].hist(df['Tool wear [min]'], bins=50, edgecolor='black')
axes[2].set_title('Tool Wear')
axes[2].set_xlabel('Minutes')

plt.tight_layout()
plt.show()
```

```python
# Compare failures vs no failures
failure_yes = df[df['Machine failure'] == 1]
failure_no = df[df['Machine failure'] == 0]

# Temperature comparison
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Process temperature
axes[0].boxplot([failure_no['Process temperature [K]'], 
                 failure_yes['Process temperature [K]']], 
                labels=['No Failure', 'Failure'])
axes[0].set_title('Process Temperature')
axes[0].set_ylabel('Temperature (K)')
axes[0].grid(True, alpha=0.3)

# Torque
axes[1].boxplot([failure_no['Torque [Nm]'], 
                 failure_yes['Torque [Nm]']], 
                labels=['No Failure', 'Failure'])
axes[1].set_title('Torque')
axes[1].set_ylabel('Nm')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"Avg Process Temp (No Failure): {failure_no['Process temperature [K]'].mean():.2f} K")
print(f"Avg Process Temp (Failure): {failure_yes['Process temperature [K]'].mean():.2f} K")
```
